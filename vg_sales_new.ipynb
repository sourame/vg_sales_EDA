9. Group the sales data of each region on Year_of_Release and plot the same (It should show which region sales was at peak on which year of release) 

```python
numeric_col_df = vg_sales_df[['Year_of_Release','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']]
sales_by_reg = numeric_col_df.iloc[:,:-1].groupby(['Year_of_Release']).sum()
sales_by_reg.reset_index(inplace=True)
sales_by_reg
```

```python
#sales_by_reg.plot(x='Year_of_Release',y=['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales'],kind="line")
sales_reg_df = sales_by_reg.melt('Year_of_Release', var_name='Region', value_name='Sales')
plt.figure(figsize=(25,10))
sns.catplot(x="Year_of_Release",y="Sales", hue='Region', 
            data=sales_reg_df, kind='point',height=5, aspect=2)
plt.xticks(rotation=90, fontsize=9)
plt.title("Sales per Region by Year of Release")
```

10. Group North American, Japan, European & Other region sales data as per Genre and plot it (It should show highest of region according to game category , ex- NA_Sales is highest in Adventure category , plot kind should be bar) 

```python
genre_df = vg_sales_df[['Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
sales_by_genre = genre_df.groupby(['Genre']).max()
sales_by_genre.reset_index(inplace=True)
sales_by_genre
```

```python
#sales_genre_df = pd.melt(sales_by_genre,id_vars=['Genre'],value_vars=['NA_Sales','EU_sales','JP_Sales','Other_Sales','Global_Sales'])
sales_genre_df = sales_by_genre.melt('Genre', var_name='Region', value_name='Sales')
plt.figure(figsize=(6,10))
sns.set(rc={"font.style":"normal",
            "axes.facecolor":(0.9, 0.9, 0.9),
            "figure.facecolor":'white',
            'axes.labelsize':12,
            'xtick.labelsize':8,
            'ytick.labelsize':8})
sns.catplot(x="Sales",y="Genre", hue='Region', 
            data=sales_genre_df, kind='bar')
plt.title("Sales per Region by Genre")
```

11. Group North American, Japan, European & Other region sales data as per Platform and plot it (Should show sales as per "Platform genre" , plot kind should be bar )

```python
pltfm_df = vg_sales_df[['Platform','NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
sales_by_pltfm = pltfm_df.groupby(['Platform']).max()
sales_by_pltfm.reset_index(inplace=True)  #sum()
sales_pltfm_df = sales_by_pltfm.melt('Platform', var_name='Region', value_name='Total_Sales')
```

```python
plt.figure(figsize=(25,10))
sns.catplot(x="Platform", y="Total_Sales", hue='Region', data=sales_pltfm_df, kind='bar',height=5, aspect=2)
plt.xticks(rotation=30, fontsize=9)
plt.show()
```

Additional points:

1. How many video games have been released for each platform?

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Platform',data=vg_sales_df,order=vg_sales_df.Platform.value_counts().index)
plt.xticks(rotation=30, fontsize=9)
plt.show()
```

2. Global Sales by Platform

```python
plt.figure(figsize=(12,8))
plt.title("Total of Global Sales of Video Games by Platform")
sns.barplot(data=vg_sales_df.groupby('Platform', as_index=False).sum(), 
            x='Global_Sales', y='Platform', 
            order=vg_sales_df.groupby('Platform', as_index=False).sum()
                .sort_values('Global_Sales', ascending=False).Platform)
```

How many video games of each genre are there in the dataset?

```python
plt.figure(figsize=(12,6))
plt.title("Count of Video Games by Genre")
sns.countplot(x='Genre', data=vg_sales_df, order=vg_sales_df.Genre.value_counts().index)
```

Global sales by genre

```python
plt.figure(figsize=(12,6))
plt.title("Total Global Sales of Video Games by Genre")
sns.barplot(data=vg_sales_df.groupby('Genre', as_index=False).sum(), y='Global_Sales', x='Genre', 
            order=vg_sales_df.groupby('Genre', as_index=False).sum().sort_values('Global_Sales', ascending=False).Genre)
```

Top 10 publishers

```python
plt.figure(figsize=(12,5))
plt.title("Count of Top 10 Publishers")
sns.countplot(x='Publisher', data=vg_sales_df, order=vg_sales_df.Publisher.value_counts().index[:10])
```

What are the Top 10 Publishers for Global_Sales in general?

```python
top_10_pub = vg_sales_df.groupby("Publisher").sum()[["Global_Sales"]].sort_values(['Global_Sales'],ascending=False).head(10)
plt.figure(figsize=(12,5))
ax = sns.scatterplot(
    data = top_10_pub.reset_index(),
    x= 'Publisher',
    y = 'Global_Sales',
    hue= 'Publisher'    
)
ax.set(xlabel='Top 10 Publishers', ylabel='Gobal Sales')
plt.xticks(rotation=90)
plt.show()
# vg_sales_df.groupby("Publisher")["Global_Sales"].sum().sort_values(ascending=False).head(10).plot.bar(figsize=(15, 5),ylabel="Global Sales in millions of copies")
```
